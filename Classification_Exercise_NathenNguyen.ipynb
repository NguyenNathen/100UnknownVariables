## Data processing

I will show how I handle the processing as it arises, but in the end I will wrap up everything to be streamlined. 


So we can see that x7 is a series of floats and scientific notation numbers in a string which represent a percentage. We can use `.replace()` to remove the `%` symbol. Scaling can be taken care of later with the rest of the numerical columns. x19 is more simple, a str representing money. We can remove the `$` symbol using `.replace` as well, and convert into floats.

For generalization we can do this with a single function, as using `.replace` for the `%` symbol will have no affect on x19 and the same goes for `$` and x7

```python
convert_num_cols = ['x7', 'x19']

def num_convert(input_str):
    # Apply to df column.
    # for column x7 and x19
    # Replace the smybols, then convert to a float
    val = float(input_str.replace('%','').replace('$',''))
    return val

def map_num_convert(dataframe, convert_cols):
    # iter over the categorical columns that 
    # need to be converted to float
    for i_conv in convert_cols:
        dataframe[i_conv] = dataframe[i_conv].apply(num_convert)
        
    return dataframe
```

```python
train_df = map_num_convert(train_df, convert_num_cols)

print(train_df['x7'].dtype)
print(train_df['x19'].dtype)
```

As previously mentioned, we have to address x3; the day of the week column. By simply taking the first three characters of any input string, we can sort everything into its corresponding day.

Example:`'Thursday'[:3]` and `'Thur'[:3]` both return `'Thu'`

```python
def day_sort(dataframe):
    # Apply to df column. Use only the first 3 letters of day strings.
    dataframe['x3'] = dataframe['x3'].apply(lambda x: x[:3]) 
    return dataframe
```

```python
train_df = day_sort(train_df)

train_df['x3'].value_counts()
```

Now that we've addressed the peculiar columns, now we can address the NaN values in the categorical data. Converting NaN values to a string and as an additional state.  

```python
def NaN_replace(input_str):
    if pd.isna(input_str):
        return str(input_str)
    else:
        return input_str
    
    
# pandas has a general .fillna command, but by doing it through .iteritems 
# it allows us to select all the object columns in the df and apply only to them
# without explicity listing out the desired columns. 
# This is just a preference of mine.     
def cat_NaN_replace(dataframe):
    for key, val in dataframe.select_dtypes('object').iteritems():
        dataframe[key] = dataframe[key].apply(NaN_replace)
        
    return dataframe
        
```

```python
train_df = cat_NaN_replace(train_df)
train_df.select_dtypes('object')
```

```python
def to_bool(input_str):
    # Apply to df column.
    # converts yes to 1, no/nan to 0
    if input_str.lower() == 'yes':
        return 1
    else:
        return 0 


def convert_cat_to_bool(dataframe):
    # Selects out those object columns with only 2 values.
    for col in dataframe.select_dtypes('object'):
        if dataframe[col].nunique() == 2:
            dataframe[col] = dataframe[col].apply(to_bool)
            
    return dataframe
```

As mentioned eariler there are quite a few columns missing data. Some have up to ~25% and some outliers even up to >75%. There are quite a few opinions on missing data, with some believing you should throw out a column if it's missing more than 30% of samples, while others believe that more data is always better. 

It's hard to come to a decision on the dataset we are working with since we don't know anything about the variables with missing values. For classification, more data always helps. We can convert columns with large amounts of missing values to boolean data. 1 if there is a measurement, 0 if not. We can set a threshold to decide what percetange of missing data we replace the column with boolean data. 

```python
def float_to_bool(input_val):
    # Converts data to bool. 1 if data, 0 if NaN
    if pd.isna(input_val):
        return 0
    else:
        return 1

def missing_convert(dataframe, threshold=0.3, conv_cols = False):
    """
    Converts columns with a given threshold of missing values to boolean data.
    
    Variables
    #########
    dataframe - pd.Dataframe
        the dataframe to apply the conversion on
    threshold - float
    conv_cols - False/list
        The columns of the traing data which do not meet the threshold. Given so that
        the same columns that the training data converts are also used in the test data. 
    
    Returns
    #########
    dataframe - pd.Dataframe
        The dataframe with converted columns.
    conv_cols - list
        List of columns which had conversion applied.
    """
    
    if conv_cols == False:
        conv_cols = []
        df_size = len(dataframe.index)
        for col in dataframe.select_dtypes('float'):
            check_val = dataframe[col].isna().sum()
            if check_val >= (df_size * threshold):           
                conv_cols.append(col)
            
    elif type(conv_cols) == list:
        conv_cols = conv_cols
        
    else:
        print('Please input a float for a threshold percent or a list of columns to convert')
        return
        
    for col in conv_cols:
        dataframe[col] = dataframe[col].apply(float_to_bool)

#     dataframe = dataframe.drop(columns=bad_cols)
            
    return dataframe, conv_cols
```

```python
convert_num_cols = ['x7', 'x19']
drop_cols = ['x39', 'x28']

def dfPrep(dataframe, 
           convert_num_cols=convert_num_cols,
          drop_cols=drop_cols, conv_cols=False, Train=True):
    
    """
    Prep the data before we put it through an sklearn pipeline.
    Apply all the cleaning we've gone through so far.
    
    Variables
    #########
    dataframe - pd.DataFrame
        The dataframe we wish to clean
    convert_num_cols - list
        Categorical columns which need to be converted to numerical data
    drop_cols - list
        Columns which we want to drop from the dataframe
    Train - bool
        True if training data, which will drop the y column as well as give the conv_cols from 
        missing_convert()
        
    Returns
    ########
    dataframe - pd.DataFrame
        Cleaned dataframe.
    
    If Train=True, will also return
    y_array - np.array
        The y values for the training data
    conv_cols - list
        Columns that were converted to binary
    
    
    """

    
    # copy the dataframe so all the actions don't alter the 
    # input dataframe, and so the output is a distinct dataframe
    dataframe=dataframe.copy()
    
    
    # iterating over the converted numerical columns
    # applying the covnersion to each column
    dataframe = map_num_convert(dataframe, convert_num_cols)
        

    # converting each day of the week into a single str
    dataframe = day_sort(dataframe)
    # replacing all NaNs in the non-numerical columns with str 'None'
    dataframe = cat_NaN_replace(dataframe)

    # convert the object columns that are boolean to int
    dataframe = convert_cat_to_bool(dataframe)
    
    # convert columns with large amounts of missing data to boolean
    dataframe, conv_cols = missing_convert(dataframe, threshold=0.3, conv_cols=conv_cols)

    # drop columns which are harmful to our models
    dataframe = dataframe.drop(columns=drop_cols)

    if Train:
        y_array = dataframe['y'].to_numpy()
        dataframe.drop(columns=['y'], inplace=True)
        return dataframe, y_array, conv_cols
    
    else:
        return dataframe
```

For our numerical columns, we want to replace the missing values and scale the data. Scaling the data gives better performance for machine learning algorithms. Scikit-learn has the `SimpleImputer` class, which will replace NaN values with a specified strategy. Let's replace our values with the median of each column. 
`SimpleImputer(strategy="median")`

`StandardScaler()` will `fit` a mean and standard diviation to each feature column. Calling `transform` will remove the mean from the data and scale to the unit variance. 

Using sklearn `Pipeline` will apply this strategy to each column of the dataset.

```python
# define ours columns as 

num_cols = train_df.select_dtypes('float').columns.tolist()
##### binary_cols = train_df.select_dtypes('int').columns.tolist()[1:] # the first in the list is the y column
##### This should have been removed
binary_cols = train_df.select_dtypes('int').columns.tolist()
categorical_cols = train_df.select_dtypes('object').columns.tolist()

# import categorical encoders form sklearn
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder

# import sklearn pipeline, imputer for replacement, and standardscaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.base import clone


# define a sub-pipeline to handle the numeric data
num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

# def the full pipeline.
full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('binary', OrdinalEncoder(), binary_cols),
      ('cat', OneHotEncoder(), categorical_cols) # forgot to call drop_first=True
])
```

## Logistic Regression

First let's define the label list, put the data through the pipeline, then split them into train and validation datasets with `train_test_split` (which is shuffled randomly through the function). We want to fit our models on the 'training' dataset, and check the the auc score with the validation dataset. So we can see how our models score data they haven't seen before. 

One final thing we should look at before we choose how to split our data and begin training is to see how our dataset is distributed by class. 

```python
bar_labels = ['0','1']
# count the amount of 0 and 1 labels, put into a corresponding list
bar_counts = [np.sum(train_labels==0), np.sum(train_labels==1)]
plt.bar(bar_labels, bar_counts)
plt.show()
```

Only ~14 percent of the data is class 1! This is an issue, we should implement a stratified split in `train_test_split`, as well as make sure we fit our models with balanced class weights which will compute class weights as `n_samples / (n_classes * np.bincount(y))` so that extra emphasis is placed on the class 1 samples.

`trian_test_split` works by randomly splitting our dataset into two new datasets while. For this case, let's make the valid dataset 5000 of the total dataset. Using `stratify` will make sure that the proportion of classes in each split is equal. 

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(train_data, train_labels, test_size=0.1,
                                                     stratify=train_labels)
```

```python
print(np.sum(y_train==1)/len(y_train))
print(np.sum(y_valid==1)/len(y_valid))
```

Great, the proportion of class 1 samples is the same in each dataset split despite being different sizes. 

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
# to get to auc score

from sklearn.linear_model import LogisticRegression
#import the logisitc regressor
# run the default solver lbfgs
# and imoprtantly, remember to run 'balanced' class weights
log_clf = LogisticRegression('l2',solver='lbfgs',
             max_iter=1000,random_state=8, class_weight='balanced')
# fit the data
log_clf.fit(X_train, y_train)
# give predictions on the validation dataset
y_predict = log_clf.predict(X_valid)

# get the auc score
print('AUC: ',roc_auc_score(y_valid, y_predict))
print('Accuracy: ', accuracy_score(y_valid,y_predict))
```

With the LogisticRegressor model, we actually achieve a decent AUC score given the dataset size and amount of features.

Beacause of the large amount of features in the dataset, a non-GLM algroithm I believe will work will is a `DecisionTree` classifier. In particular, let's use a boosting method. The popular `AdaBoost` 
( Adaptive Boosting) should work well. 

But there are quite a few hyperparameters such as `learning_rate` and `n_estimators`. It would be good to see which hyperparameters would give us the best performance. We can use `GridSearchCV` to go through a a grid of parameters. Running cross validation on the given dataset for a set of parameters. Cross validation in `sklearn` means to split a random part of the dataset off as a cross validation set, and fit the model with the remaining model then use the cross validation set to compute a score. 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

# Defining our grid of hyperparameters
param_grid = [
    # try 27 (3×3x3) combinations of hyperparameters
    {'n_estimators':[200,300,400],
     'learning_rate':[0.4,0.5,0.6,0.7,0.8],
     'base_estimator__max_depth':[1]
    }

]

# define the model again, with the parameters in the grid not defined.

ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(class_weight='balanced'),
    algorithm="SAMME.R")

# Timing the processes
start_time = time.time()
# defining gridsearch, with roc_auc score. 
grid_search = GridSearchCV(ada_clf, param_grid, cv=5,
                           scoring='roc_auc',
                           return_train_score=True)
# fitting the model
grid_search.fit(X_train, y_train)
gridsearch_time = time.time() - start_time
# print the best paramaters. 
grid_search.best_params_
```

```python

# Call the AdaBoost Classifier
# and the DecisionTreeClassifier as the base estimator.
# The default algorithm SAMME.R 
# Again, remember to call balanced weights. 
ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1,class_weight='balanced'),
    n_estimators=200,
    algorithm="SAMME.R", learning_rate=0.5)
# fit the model
ada_clf.fit(X_train, y_train)
# give predictions on the validation dataset
y_pred = ada_clf.predict(X_valid)
#get the auc score
print('AUC: ',roc_auc_score(y_valid, y_pred))
print('Accuracy: ', accuracy_score(y_valid,y_pred))
```

So our AdaBoost classifier gives a slightly better performance than our Logisitc Regressor. As these are 100 unknown variables, and a linear correlation is not certin, I believe these results did well for both models.
